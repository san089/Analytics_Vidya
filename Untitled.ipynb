```python
a = re.sub(pattern='#(\w+)' , repl='' ,string='@peter I really love that shirt at #Macy. http://bet.ly//WjdiW4' )
a = re.sub(pattern='@(\w+)' , repl='' ,string= a )
a = re.sub(pattern='http?:[a-zA-Z.//0-9]+' , repl='' , string=a)
a = re.sub(pattern='[^0-9a-zA-Z ]' , repl='' , string=a)
```

```python
def clean_data(a):
    #a = re.sub(pattern='#(\w+)' , repl='' ,string=a)
    a = re.sub(pattern='@(\w+)' , repl='' ,string= a )
    a = re.sub(pattern='(http(|s):[a-zA-Z.-//0-9]+)|(#(\w+))' , repl='' , string=a)
    a = re.sub(pattern='[^0-9a-zA-Z.\' \$]' , repl='' , string=a)
    return a
```

```python
tweet_df['clean_tweet'] = tweet_df['tweet'].map(clean_data)
```
